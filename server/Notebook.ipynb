```python
import pandas as pd
from tinydb import TinyDB, Query
```

```python
db = TinyDB('.\\db.json')
```

```python
db.all()[0].doc_id
```

```python
data = [['Bilal Khan', 'Software Engineer', 'Laique Hassan', 85],
        ['Ammar Baig', 'Software Engineer', 'Laique Hassan', 85],
        ['Laique Hassan', 'Senior Software Engineer', 'Adnan Hirani', 85],
        ['Adnan Hirani', 'Engineering Manager', 'Tauqeer Iqbal', 85],
        ['Tauqeer Iqbal', 'Senior Engineering Manager', 'Salman Kasbati', 85]]
df = pd.DataFrame(columns=['name', 'position', 'manager', 'allocation'], data=data)
df
```

```python
teams = db.table('teams')
```

```python
for index, row in df.iterrows():
    teams.insert({'name': row[0], 'department': row[1], 'reporting': row[2], 'allocation': row[3], 'projects': row[4]})

teams.all()
```
